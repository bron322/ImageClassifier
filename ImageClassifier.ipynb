# Data Preprocessing
## Scale the data

```python
data = data.map(lambda x,y: (x/255, y))
```

https://www.tensorflow.org/api_docs/python/tf/data/Dataset#map

```python
data.as_numpy_iterator().next()
```

## Splitting the data

```python
len(data)
```

```python
train_size = int(len(data)*.7)
val_size = int(len(data)*.2)
test_size = int(len(data)*.1)
```

```python
train_size
```

```python
train = data.take(train_size)
val = data.skip(train_size).take(val_size)
test = data.skip(train_size+val_size).take(test_size)
```

# Deep Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
```

```python
model = Sequential()
```

```python
model.add(Conv2D(16, (3,3), 1, activation='relu', input_shape=(256,256,3)))
model.add(MaxPooling2D())
model.add(Conv2D(32, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())
model.add(Conv2D(16, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

`relu` function means any negative values will be converted to 0 and we preserve the positive values

`16`: This specifies the number of filters (also known as kernels or channels) in the convolutional layer. Each filter detects different features in the input data.
`(3,3)`: This specifies the size of the convolutional kernel or filter. In this case, it's a 3x3 filter, meaning it will convolve over a 3x3 region of the input at a time.
`1`: This specifies the stride of the convolution operation. A stride of 1 means the filter moves one pixel at a time.

The reason why `sigmoid` activation is chosen here is because the function effectively turns the output between 0 to 1. Remember 0 and 1 map to our Happy and Sad classes

```python
model.compile('adam', loss=tf.losses.BinaryCrossentropy(), metrics=['accuracy'])
```

```python
model.summary()
```

Note that we are condensing the MaxPooling2D layer which is 30x30x16 into flatten layer which is 14400

# Training the model

```python
logdir='logs'
```

```python
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=logdir)
tensorboard_callback 
```

```python
hist = model.fit(train, epochs=20, validation_data=val, callbacks=[tensorboard_callback])
```

```python
hist.history
```

## Plotting the performance

```python
fig = plt.figure()
plt.plot(hist.history['loss'], color='teal', label='loss')
plt.plot(hist.history['val_loss'], color='orange', label='val_loss')
fig.suptitle('Loss', fontsize=20)
plt.legend(loc="upper left")
plt.show()
```

```python
fig = plt.figure()
plt.plot(hist.history['accuracy'], color='teal', label='accuracy')
plt.plot(hist.history['val_accuracy'], color='orange', label='val_accuracy')
fig.suptitle('Accuracy', fontsize=20)
plt.legend(loc="upper left")
plt.show()
```

# Evaluate

```python
from tensorflow.keras.metrics import Precision, Recall, BinaryAccuracy
```

```python
pre = Precision()
re = Recall()
acc = BinaryAccuracy()
```

```python
for batch in test.as_numpy_iterator(): 
    X, y = batch
    yhat = model.predict(X)
    pre.update_state(y, yhat)
    re.update_state(y, yhat)
    acc.update_state(y, yhat)
```

```python
print(f'Precision:{pre.result().numpy()}, Recall:{re.result().numpy()}, Accuracy:{acc.result().numpy()}')
```
